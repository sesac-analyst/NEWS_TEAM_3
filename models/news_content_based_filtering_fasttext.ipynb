# Tokenization

```python
from gensim.models import FastText
from ekonlpy.tag import Mecab
import pandas as pd
```

```python
mecab = Mecab()

# 뉴스 데이터프레임에서 content 컬럼을 토크나이징
def tokenize_content(text):
    return mecab.morphs(text) 
```

```python
# 각 content를 토큰화한 결과를 저장
news_df.loc[:, 'tokenized_content'] = news_df['content'].apply(tokenize_content)
```

```python
news_df.head(2)
```

```python
# 모든 토큰화된 content를 리스트 형태로 변환하여 모델에 학습
sentences = news_df['tokenized_content'].tolist()
```

```python
# FastText 모델 학습
model = FastText(sentences=sentences, vector_size=100, window=5, min_count=1, workers=4, sg=1, epochs=10)
```

```python
word_vector = model.wv['경제']
word_vector
```

```python
# 각 문서의 단어 임베딩 벡터 평균을 계산하여 문서 임베딩 벡터 생성
def get_document_embedding(tokens, model):
    # 토큰이 모델에 존재하는 경우 임베딩 벡터를 가져와서 평균 계산
    return sum([model.wv[token] for token in tokens if token in model.wv]) / len(tokens)

news_df['content_embedding'] = news_df['tokenized_content'].apply(lambda x: get_document_embedding(x, model))
```

```python
def get_similar_articles(article_url, top_n=10):
    # 입력받은 title과 일치하는 기사의 인덱스 찾기
    idx = sampled_news_df[sampled_news_df['title'] == article_url].index[0]
    
    # 해당 기사의 유사한 데이터 인덱스 가져오기
    sim_scores = cosine_sim_matrix[idx]
    
    # 유사도 순으로 정렬된 상위 top_n개의 기사 인덱스 가져오기
    similar_idx = np.argsort(-sim_scores)[1:top_n+1]  # 자기 자신을 제외하고 [1:]부터
    
    # 반환
    return sampled_news_df.iloc[similar_idx][['title', 'content', 'article_url']]
```
