```python
# 각 문서의 단어 임베딩 벡터 평균을 계산하여 문서 임베딩 벡터 생성
def get_document_embedding(tokens, model):
    # 토큰이 모델에 존재하는 경우 임베딩 벡터를 가져와서 평균 계산
    return sum([model.wv[token] for token in tokens if token in model.wv]) / len(tokens)

news_df['content_embedding'] = news_df['tokenized_content'].apply(lambda x: get_document_embedding(x, model))
```

```python
def get_similar_articles(article_url, top_n=10):
    # 입력받은 title과 일치하는 기사의 인덱스 찾기
    idx = sampled_news_df[sampled_news_df['title'] == article_url].index[0]
    
    # 해당 기사의 유사한 데이터 인덱스 가져오기
    sim_scores = cosine_sim_matrix[idx]
    
    # 유사도 순으로 정렬된 상위 top_n개의 기사 인덱스 가져오기
    similar_idx = np.argsort(-sim_scores)[1:top_n+1]  # 자기 자신을 제외하고 [1:]부터
    
    # 반환
    return sampled_news_df.iloc[similar_idx][['title', 'content', 'article_url']]
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# 임베딩된 벡터를 나누어 처리하는 함수
def batched_cosine_similarity(vectors, batch_size):
    num_vectors = vectors.shape[0]
    sim_matrix = np.zeros((num_vectors, num_vectors))

    for i in range(0, num_vectors, batch_size):
        for j in range(i, num_vectors, batch_size):
            # 배치로 코사인 유사도를 계산
            sim = cosine_similarity(vectors[i:i + batch_size], vectors[j:j + batch_size])
            sim_matrix[i:i + batch_size, j:j + batch_size] = sim
            sim_matrix[j:j + batch_size, i:i + batch_size] = sim.T  # 대칭성 활용
    return sim_matrix

# 1. 이미 FastText로 임베딩된 content_embedding을 NumPy 배열로 변환
document_embeddings_array = np.array(news_df['content_embedding'].tolist())

# 2. 배치 크기를 설정하고 코사인 유사도를 계산
batch_size = 1000
cosine_sim_matrix = batched_cosine_similarity(document_embeddings_array, batch_size)

# 코사인 유사도 행렬 출력
print(cosine_sim_matrix)
```

# Evaluation

```python
from sklearn.model_selection import train_test_split

# news_df 데이터프레임을 학습과 테스트로 분리
train_df, test_df = train_test_split(news_df, test_size=0.2, random_state=42)

# 학습 데이터: 유저가 읽은 기사 정보 (content_embedding 열은 이미 FastText로 임베딩된 벡터)
train_embeddings = np.array(train_df['content_embedding'].tolist())

# 테스트 데이터: 유저가 새롭게 읽을 수 있는 기사 정보 (예측을 위한 데이터)
test_embeddings = np.array(test_df['content_embedding'].tolist())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend_articles(user_embedding, article_embeddings, top_n=5):
    # 유저 벡터와 기사 벡터들 간의 코사인 유사도를 계산
    sim_scores = cosine_similarity(user_embedding.reshape(1, -1), article_embeddings)
    
    # 유사도 순으로 상위 top_n개의 기사 인덱스 반환
    top_indices = np.argsort(-sim_scores)[0][:top_n]
    
    return top_indices

# 예시로 테스트 데이터의 첫 번째 기사를 기준으로 추천
user_embedding = test_embeddings[0]  # 임의의 유저 벡터 (테스트 데이터에서 선택)
recommended_indices = recommend_articles(user_embedding, train_embeddings)

# 추천된 기사들을 출력 (train_df에서 추천된 기사의 제목과 내용 가져오기)
recommended_articles = train_df.iloc[recommended_indices][['title', 'content']]
recommended_articles
```

```python
# 추천된 기사가 실제로 유저가 읽은 기사인지 확인하는 함수
def evaluate_precision(test_df, recommended_indices, top_n=5):
    # 테스트 데이터에서 유저가 실제로 읽은 기사 확인 (임의의 유저로 진행)
    actual_articles = set(test_df['title'])  # 실제 유저가 읽은 기사 (테스트 데이터에 있는 기사들)
    
    # 추천된 기사 중에서 실제로 유저가 읽은 기사와 일치하는 개수 확인
    recommended_articles = train_df.iloc[recommended_indices]['title'].values
    
    relevant_articles = [article for article in recommended_articles if article in actual_articles]
    
    # Precision 계산: 추천된 기사 중 실제로 유저가 읽은 기사의 비율
    precision = len(relevant_articles) / top_n
    return precision

# 평가 함수 실행 (테스트 유저의 기사와 비교)
precision = evaluate_precision(test_df, recommended_indices, top_n=5)
print(f"Precision: {precision:.2f}")
```
